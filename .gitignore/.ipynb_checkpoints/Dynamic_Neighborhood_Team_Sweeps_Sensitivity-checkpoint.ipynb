# Dynamic Neighborhood — Team Optimum

This notebook analyzes the **team-optimum** solution for the **dynamic neighborhood** model.

Workflow:

1. import the evolving-neighborhood model and define notebook-wide settings,
2. run parameter sweeps for the team optimum,
3. generate summary plots for the converged values of $N$, $u$, and $y$,
4. compute local sensitivity analysis around a baseline parameter set.

This version is streamlined to avoid repeated imports, repeated setup code, and repeated plotting logic.


```python
# ============================================================
# Imports and notebook-wide setup
# ============================================================
from pathlib import Path
from dataclasses import replace
from typing import Sequence, Optional

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# Project imports
from utils.model_params import EvolvingParams
from utils.evolving_neighborhood import integrate_teamopt, G

# ------------------------------------------------------------
# Output directories
# ------------------------------------------------------------
PROJECT_ROOT = Path(".")
OUTDIR = PROJECT_ROOT / "outputs" / "evolving_neighborhoods" / "team_opt_dynamics"
OUTDIR.mkdir(parents=True, exist_ok=True)

SWEEP_DIR = OUTDIR / "parameter_sweeps"
SWEEP_DIR.mkdir(parents=True, exist_ok=True)

SUMMARY_DIR = SWEEP_DIR / "summary_plots"
SUMMARY_DIR.mkdir(parents=True, exist_ok=True)

SENS_DIR = OUTDIR / "sensitivity"
SENS_DIR.mkdir(parents=True, exist_ok=True)

print("Sweep outputs:", SWEEP_DIR.resolve())
print("Sensitivity outputs:", SENS_DIR.resolve())
```

## Baseline model parameters

The dynamic-neighborhood model uses the `EvolvingParams` parameter dataclass.
The neighborhood size is itself dynamic, so the state variables are $(N,u,y)$.

We define a baseline parameter object here and then override individual parameters during sweeps or sensitivity analysis.


```python
# ============================================================
# Baseline parameters and initial conditions
# ============================================================
BASE_P = replace(
    EvolvingParams(),
    t_max=50_000.0,
    rtol=1e-5,
    atol=1e-7,
    max_step=5.0,
    N_floor=1e-9,
    u_eps=1e-10,
)

# Initial conditions
N0_default = 5.0
u0_default = 0.5
y0_default = 0.5

BASE_P
```

## Sweep helper functions

The helper functions below do the following:

- safely format parameter values for filenames and Excel sheet names,
- enforce non-negativity for plotting and summary extraction,
- run a single team-optimum simulation,
- save each full sweep to Excel,
- generate per-run diagnostic trajectory plots,
- build one compact summary plot per swept parameter.


```python
# ============================================================
# Helper functions for parameter sweeps
# ============================================================
def _tag(v) -> str:
    """Convert numeric values into filename-safe tags."""
    if isinstance(v, (int, np.integer)):
        return str(int(v))
    s = f"{float(v):.6g}"
    return s.replace("-", "m").replace(".", "p")


def _untag_value(token: str) -> float:
    """Reverse of _tag()."""
    s = token.replace("p", ".")
    if s.startswith("m"):
        s = "-" + s[1:]
    return float(s)


def _truncate_at_first_negative(arr: np.ndarray) -> np.ndarray:
    """If a trajectory becomes negative, set that point and all later points to zero."""
    arr = np.asarray(arr, dtype=float)
    neg = arr < 0
    if neg.any():
        k = int(np.argmax(neg))
        out = arr.copy()
        out[k:] = 0.0
        return out
    return arr


def _enforce_nonneg(N: np.ndarray, u: np.ndarray, y: np.ndarray, mode: str = "truncate"):
    """Apply non-negativity handling to the simulated trajectories."""
    N = np.maximum(N, 0.0)
    if mode == "truncate":
        u = _truncate_at_first_negative(np.maximum(u, 0.0))
        y = _truncate_at_first_negative(np.maximum(y, 0.0))
    elif mode == "clip":
        u = np.maximum(u, 0.0)
        y = np.maximum(y, 0.0)
    else:
        raise ValueError("mode must be 'truncate' or 'clip'")
    return N, u, y


def run_teamopt_evolving(
    p: EvolvingParams,
    N0: float = N0_default,
    u0: float = u0_default,
    y0: float = y0_default,
    nonneg_mode: str = "truncate",
):
    """Run one team-optimum simulation for the evolving-neighborhood model."""
    out = integrate_teamopt(q=0.0, p=p, N0=N0, u0=u0, y0=y0)

    if not out["success"]:
        print(f"[warn] LSODA: {out['message']}")

    t = np.asarray(out["t"], dtype=float)
    N = np.asarray(out["N"], dtype=float)
    u = np.asarray(out["u"], dtype=float)
    y = np.asarray(out["y"], dtype=float)

    N, u, y = _enforce_nonneg(N, u, y, mode=nonneg_mode)
    return t, N, u, y


def _plot_three_panel_teamopt(
    t,
    N,
    u,
    y,
    p: EvolvingParams,
    title_extra: str,
    figpath_base: Optional[Path] = None,
):
    """Plot N(t), u(t), and y(t) for one team-optimum run."""
    fig, axs = plt.subplots(1, 3, figsize=FIGSIZE_RUN)

    axs[0].plot(t, N, linewidth=LINEWIDTH, color=COLOR_N)
    axs[0].set_title("Neighborhood size N(t)")
    axs[0].set_xlabel("Time")
    axs[0].set_ylabel("N")
    axs[0].grid(True, alpha=0.25)

    axs[1].plot(t, u, linewidth=LINEWIDTH, color=COLOR_U)
    axs[1].set_title("VEGF u(t) — Team optimum")
    axs[1].set_xlabel("Time")
    axs[1].set_ylabel(r"$u_{\mathrm{team}}$")
    axs[1].grid(True, alpha=0.25)

    axs[2].plot(t, y, linewidth=LINEWIDTH, color=COLOR_Y)
    axs[2].set_title("GLUT1 y(t) — Team optimum")
    axs[2].set_xlabel("Time")
    axs[2].set_ylabel(r"$y_{\mathrm{team}}$")
    axs[2].grid(True, alpha=0.25)

    fig.tight_layout(rect=[0, 0, 1, 0.92])
    fig.suptitle(
        f"Team-opt dynamics — {title_extra} | "
        f"R={p.R}, a={p.a}, k={p.k}, c={p.c}, h={p.h}, f={p.f}, t_max={int(p.t_max)}",
        y=0.99,
        fontsize=16,
    )

    if figpath_base is not None:
        fig.savefig(figpath_base.with_suffix(".png"), bbox_inches="tight")
        fig.savefig(figpath_base.with_suffix(".svg"), bbox_inches="tight")

    plt.show()


def sweep_teamopt_evolving(
    param: str,
    values: Sequence[float],
    N0_init: float = N0_default,
    u0: float = u0_default,
    y0: float = y0_default,
    b_override: Optional[float] = None,
    sigma_u_override: Optional[float] = None,
    sigma_y_override: Optional[float] = None,
    tmax_override: Optional[float] = None,
):
    """Sweep one parameter and save the full trajectories to Excel."""
    if not hasattr(BASE_P, param):
        raise AssertionError(f"Unknown parameter '{param}'")

    param_dir = SWEEP_DIR / param
    param_dir.mkdir(parents=True, exist_ok=True)

    excel_path = param_dir / f"teamopt_evo_sweep_{param}_N0{_tag(N0_init)}.xlsx"
    print(f"\nRunning team-optimum sweep for '{param}'")
    print(f"Saving workbook to: {excel_path}")

    all_runs = {}

    with pd.ExcelWriter(excel_path, engine="openpyxl") as writer:
        pd.DataFrame({
            "Parameter": [param],
            "Values": [list(map(float, values))],
            "N0_init": [N0_init],
            "u0": [u0],
            "y0": [y0],
            "b_override": [b_override],
            "sigma_u_override": [sigma_u_override],
            "sigma_y_override": [sigma_y_override],
            "tmax_override": [tmax_override],
        }).to_excel(writer, sheet_name="params", index=False)

        for val in values:
            val = float(val)

            kwargs = {param: val}
            if b_override is not None:
                kwargs["b"] = b_override
            if sigma_u_override is not None:
                kwargs["sigma_u"] = sigma_u_override
            if sigma_y_override is not None:
                kwargs["sigma_y"] = sigma_y_override
            if tmax_override is not None:
                kwargs["t_max"] = tmax_override

            p_run = replace(BASE_P, **kwargs)

            t, N, u, y = run_teamopt_evolving(p_run, N0=N0_init, u0=u0, y0=y0)

            df = pd.DataFrame({"t": t, "N": N, "u": u, "y": y})
            sheet_name = f"{param}_{_tag(val)}"
            df.to_excel(writer, sheet_name=sheet_name, index=False)
            all_runs[sheet_name] = df

            figbase = param_dir / f"{param}_{_tag(val)}_N0{_tag(N0_init)}"
            _plot_three_panel_teamopt(
                t=t,
                N=N,
                u=u,
                y=y,
                p=p_run,
                title_extra=f"{param}={val:.3g}",
                figpath_base=figbase,
            )

    print(f"Finished sweep: {param}")
    return all_runs


def extract_final_teamopt_from_sweep_dict(one_sweep_result):
    """Extract final values N*, u*, y* from one sweep result dictionary."""
    rows = []

    for sheet_name, df in one_sweep_result.items():
        if df.empty:
            continue

        try:
            token = sheet_name.split("_", 1)[1]
            param_value = _untag_value(token)
        except Exception:
            continue

        rows.append({
            "parameter_value": param_value,
            "N_final": float(df["N"].iloc[-1]),
            "u_final": float(df["u"].iloc[-1]),
            "y_final": float(df["y"].iloc[-1]),
            "t_final": float(df["t"].iloc[-1]),
        })

    return pd.DataFrame(rows).sort_values("parameter_value").reset_index(drop=True)


def plot_teamopt_summary(df_final, param_name, xlabel=None, title=None, save=True):
    """Plot N*, u*, and y* across the swept parameter in a 1×3 summary figure."""
    if df_final.empty:
        print(f"[warn] No data available for {param_name}")
        return

    fig, axs = plt.subplots(1, 3, figsize=FIGSIZE_SUMMARY)
    x = df_final["parameter_value"].values

    axs[0].plot(x, df_final["N_final"].values, color=COLOR_N, linewidth=LINEWIDTH)
    axs[0].set_title(r"Final $N_{\mathrm{team}}$")
    axs[0].set_xlabel(xlabel if xlabel else param_name)
    axs[0].set_ylabel(r"$N_{\mathrm{team}}$")
    axs[0].grid(False)

    axs[1].plot(x, df_final["u_final"].values, color=COLOR_U, linewidth=LINEWIDTH)
    axs[1].set_title(r"Final $u_{\mathrm{team}}$")
    axs[1].set_xlabel(xlabel if xlabel else param_name)
    axs[1].set_ylabel(r"$u_{\mathrm{team}}$")
    axs[1].grid(False)

    axs[2].plot(x, df_final["y_final"].values, color=COLOR_Y, linewidth=LINEWIDTH)
    axs[2].set_title(r"Final $y_{\mathrm{team}}$")
    axs[2].set_xlabel(xlabel if xlabel else param_name)
    axs[2].set_ylabel(r"$y_{\mathrm{team}}$")
    axs[2].grid(False)

    if title is None:
        title = f"Final team-opt traits across {param_name}"
    fig.suptitle(title, y=0.98, fontsize=16)
    fig.tight_layout(rect=[0, 0, 1, 0.93])

    if save:
        fig.savefig(SUMMARY_DIR / f"teamopt_evo_summary_{param_name}.png", bbox_inches="tight")
        fig.savefig(SUMMARY_DIR / f"teamopt_evo_summary_{param_name}.svg", bbox_inches="tight")

    plt.show()
```

# Sensitivity

We compute **local, central-difference sensitivities** of the team-optimum equilibrium
values $(N^*,u^*,y^*)$ with respect to
$$
p \in \{k, a, R, c, h, f\}.
$$

For each parameter, we perturb the baseline value upward and downward,
integrate the dynamic-neighborhood team-optimum to equilibrium, and estimate
the local sensitivity using finite differences.


## Central-difference perturbations

Let $\alpha = \mathrm{FRAC\_STEP}$ and let $\mathrm{MIN\_ABS}[p]$ denote the
minimum perturbation size for parameter $p$. For each parameter,
$$
\Delta p = \max\!\big(\alpha |p_0|,\ \mathrm{MIN\_ABS}[p]\big), \qquad
p_- = p_0 - \Delta p, \qquad
p_+ = p_0 + \Delta p.
$$

We then estimate local derivatives by central differences and convert them into
normalized sensitivities for comparison across parameters.


```python
# ============================================================
# TEAM-OPTIMUM LOCAL SENSITIVITY (EVOLVING NEIGHBORHOOD)
# ============================================================

# ------------------------------------------------------------
# Baseline configuration for sensitivity analysis
# ------------------------------------------------------------
SENS_BASE = replace(EvolvingParams(), t_max=250_000.0)
SENS_N0 = 5.0
SENS_u0 = 0.5
SENS_y0 = 0.5

# Parameters to perturb. For the dynamic-neighborhood model,
# N is already a state variable, so we vary model parameters only.
params_of_interest = ["k", "a", "R", "c", "h", "f"]

# Finite-difference settings
FRAC_STEP = 0.10
MIN_ABS = {"k": 1e-3, "a": 1e-3, "R": 0.5, "c": 1e-3, "h": 1e-3, "f": 1e-3}
EPS = 1e-9

# Convergence settings
CHECK = dict(window_frac=0.1, atol=1e-5, rtol=1e-4)
MAX_EXTENSIONS = 6
TMAX_FACTOR = 2.0

# ------------------------------------------------------------
# Helper functions
# ------------------------------------------------------------
def has_equilibrated(t, N, u, y, window_frac=0.1, atol=1e-5, rtol=1e-4):
    """Check whether N, u, and y have stabilized over the final part of the trajectory."""
    n = len(t)
    if n < 5:
        return False

    k0 = max(0, int((1.0 - window_frac) * n))
    N_win, u_win, y_win = N[k0:], u[k0:], y[k0:]
    Nf, uf, yf = N_win[-1], u_win[-1], y_win[-1]

    N_ok = np.max(np.abs(N_win - Nf)) <= (atol + rtol * max(1.0, abs(Nf)))
    u_ok = np.max(np.abs(u_win - uf)) <= (atol + rtol * max(1.0, abs(uf)))
    y_ok = np.max(np.abs(y_win - yf)) <= (atol + rtol * max(1.0, abs(yf)))
    return bool(N_ok and u_ok and y_ok)


def integrate_to_equilibrium_team(
    p: EvolvingParams,
    N0: float = SENS_N0,
    u0: float = SENS_u0,
    y0: float = SENS_y0,
    max_extensions: int = MAX_EXTENSIONS,
    tmax_factor: float = TMAX_FACTOR,
    check_kwargs: Optional[dict] = None,
    nonneg_mode: str = "truncate",
):
    """Integrate team-optimum dynamics to equilibrium, extending t_max when needed."""
    if check_kwargs is None:
        check_kwargs = CHECK

    p_run = p
    extensions = 0

    while True:
        out = integrate_teamopt(q=0.0, p=p_run, N0=N0, u0=u0, y0=y0)

        if not out["success"]:
            print(f"[warn] LSODA: {out['message']}")

        t = np.asarray(out["t"], dtype=float)
        N = np.asarray(out["N"], dtype=float)
        u = np.asarray(out["u"], dtype=float)
        y = np.asarray(out["y"], dtype=float)
        N, u, y = _enforce_nonneg(N, u, y, mode=nonneg_mode)

        if has_equilibrated(t, N, u, y, **check_kwargs):
            return {
                "t": t,
                "N": N,
                "u": u,
                "y": y,
                "p_used": p_run,
                "converged": True,
                "extensions": extensions,
            }

        if extensions >= max_extensions:
            return {
                "t": t,
                "N": N,
                "u": u,
                "y": y,
                "p_used": p_run,
                "converged": False,
                "extensions": extensions,
            }

        p_run = replace(p_run, t_max=p_run.t_max * tmax_factor)
        extensions += 1


def _perturb_team(prm: str, base_p: EvolvingParams, frac=FRAC_STEP):
    """Generate minus and plus perturbations for one parameter."""
    p0 = getattr(base_p, prm)
    dp = max(abs(frac * p0), MIN_ABS[prm])
    pm = replace(base_p, **{prm: max(p0 - dp, 0.0)})
    pp = replace(base_p, **{prm: p0 + dp})
    return pm, pp, p0, dp


def _plot_team_diagnostics(runs: dict, title: str, save_base: Optional[Path] = None):
    """Plot diagnostic trajectories for N(t), u(t), and y(t)."""
    fig, axs = plt.subplots(1, 3, figsize=FIGSIZE_RUN)

    for lab, (t, N, _, _) in runs.items():
        axs[0].plot(t, N, lw=2.2, label=lab)
    axs[0].set_xlabel("Time")
    axs[0].set_ylabel("N(t)")
    axs[0].grid(True, alpha=0.25)
    axs[0].legend(frameon=False, fontsize=9)

    for lab, (t, _, u, _) in runs.items():
        axs[1].plot(t, u, lw=2.2, label=lab)
    axs[1].set_xlabel("Time")
    axs[1].set_ylabel(r"$u(t)$")
    axs[1].grid(True, alpha=0.25)

    for lab, (t, _, _, y) in runs.items():
        axs[2].plot(t, y, lw=2.2, label=lab)
    axs[2].set_xlabel("Time")
    axs[2].set_ylabel(r"$y(t)$")
    axs[2].grid(True, alpha=0.25)

    fig.suptitle(title, y=0.99, fontsize=15)
    fig.tight_layout(rect=[0, 0, 1, 0.95])

    if save_base is not None:
        fig.savefig(save_base.with_suffix(".png"), dpi=300)
        fig.savefig(save_base.with_suffix(".pdf"))

    plt.show()


# ------------------------------------------------------------
# Baseline equilibrium
# ------------------------------------------------------------
base_run = integrate_to_equilibrium_team(SENS_BASE, N0=SENS_N0, u0=SENS_u0, y0=SENS_y0)

N_star0 = float(base_run["N"][-1])
u_star0 = float(base_run["u"][-1])
y_star0 = float(base_run["y"][-1])
G_star0 = float(G(N_star0, u_star0, y_star0, p=base_run["p_used"]))

print(
    f"[baseline team optimum] converged={base_run['converged']}  "
    f"t_max_used={int(base_run['p_used'].t_max)}  "
    f"N*={N_star0:.5g}  u*={u_star0:.5g}  y*={y_star0:.5g}  G*={G_star0:.5g}"
)

# ------------------------------------------------------------
# Loop over perturbed parameters
# ------------------------------------------------------------
rows = []

for prm in params_of_interest:
    pm, pp, p0_val, dp_single = _perturb_team(prm, SENS_BASE, frac=FRAC_STEP)

    run_m = integrate_to_equilibrium_team(pm, N0=SENS_N0, u0=SENS_u0, y0=SENS_y0)
    run_p = integrate_to_equilibrium_team(pp, N0=SENS_N0, u0=SENS_u0, y0=SENS_y0)

    # Diagnostic plots for the baseline and ± perturbations
    diag_base = SENS_DIR / "diagnostics" / f"teamopt_evo_dyn_{prm}"
    diag_base.parent.mkdir(parents=True, exist_ok=True)

    runs = {
        f"baseline (t_max={int(base_run['p_used'].t_max)})": (base_run["t"], base_run["N"], base_run["u"], base_run["y"]),
        f"{prm}− (t_max={int(run_m['p_used'].t_max)})": (run_m["t"], run_m["N"], run_m["u"], run_m["y"]),
        f"{prm}+ (t_max={int(run_p['p_used'].t_max)})": (run_p["t"], run_p["N"], run_p["u"], run_p["y"]),
    }

    _plot_team_diagnostics(
        runs,
        title=f"Team-opt dynamics — {prm} sensitivity (dynamic neighborhood)",
        save_base=diag_base,
    )

    # Final values under minus and plus perturbations
    N_m, u_m, y_m = float(run_m["N"][-1]), float(run_m["u"][-1]), float(run_m["y"][-1])
    N_p, u_p, y_p = float(run_p["N"][-1]), float(run_p["u"][-1]), float(run_p["y"][-1])

    G_m = float(G(N_m, u_m, y_m, p=run_m["p_used"]))
    G_p = float(G(N_p, u_p, y_p, p=run_p["p_used"]))

    # Central-difference derivatives
    dp_total = 2.0 * dp_single
    dN_dp = (N_p - N_m) / dp_total
    du_dp = (u_p - u_m) / dp_total
    dy_dp = (y_p - y_m) / dp_total
    dG_dp = (G_p - G_m) / dp_total

    # Normalized sensitivities
    S_N = (p0_val / max(abs(N_star0), EPS)) * dN_dp
    S_u = (p0_val / max(abs(u_star0), EPS)) * du_dp
    S_y = (p0_val / max(abs(y_star0), EPS)) * dy_dp
    S_G = (p0_val / max(abs(G_star0), EPS)) * dG_dp

    rows.append({
        "param": prm,
        "p0": p0_val,
        "dp_single": dp_single,
        "N_star": N_star0,
        "u_star": u_star0,
        "y_star": y_star0,
        "G_star": G_star0,
        "N_minus": N_m,
        "N_plus": N_p,
        "dN/dp": dN_dp,
        "S_N": S_N,
        "u_minus": u_m,
        "u_plus": u_p,
        "du/dp": du_dp,
        "S_u": S_u,
        "y_minus": y_m,
        "y_plus": y_p,
        "dy/dp": dy_dp,
        "S_y": S_y,
        "G_minus": G_m,
        "G_plus": G_p,
        "dG/dp": dG_dp,
        "S_G": S_G,
        "tmax_used_baseline": int(base_run["p_used"].t_max),
        "tmax_used_minus": int(run_m["p_used"].t_max),
        "tmax_used_plus": int(run_p["p_used"].t_max),
        "converged_baseline": base_run["converged"],
        "converged_minus": run_m["converged"],
        "converged_plus": run_p["converged"],
    })

team_sensitivity_df = pd.DataFrame(rows).set_index("param")
team_sensitivity_df
```

## Plot normalized sensitivities

The bar plot below compares the normalized local sensitivities of
$N_{\mathrm{team}}$, $u_{\mathrm{team}}$, and $y_{\mathrm{team}}$.


```python
# ============================================================
# Plot normalized sensitivities
# ============================================================
fig, ax = plt.subplots(figsize=(9, 4.5))

x = np.arange(len(params_of_interest))
width = 0.25

S_N_vals = [team_sensitivity_df.loc[p, "S_N"] for p in params_of_interest]
S_u_vals = [team_sensitivity_df.loc[p, "S_u"] for p in params_of_interest]
S_y_vals = [team_sensitivity_df.loc[p, "S_y"] for p in params_of_interest]

ax.bar(x - width, S_N_vals, width, label=r"$N_{\mathrm{team}}$", color=COLOR_N)
ax.bar(x,         S_u_vals, width, label=r"$u_{\mathrm{team}}$", color=COLOR_U)
ax.bar(x + width, S_y_vals, width, label=r"$y_{\mathrm{team}}$", color=COLOR_Y)

ax.axhline(0, linewidth=1, color="k")
ax.set_xticks(x)
ax.set_xticklabels(params_of_interest)
ax.set_ylabel("Sensitivity")
ax.set_title("Local sensitivities at baseline (team optimum, dynamic neighborhood)")
ax.legend()
ax.grid(False)

fig.tight_layout()

fig_pdf = SENS_DIR / "teamopt_evo_local_sensitivity_bars.pdf"
fig_png = SENS_DIR / "teamopt_evo_local_sensitivity_bars.png"
fig.savefig(fig_pdf)
fig.savefig(fig_png, dpi=300)

print(f"Saved bar plot -> {fig_pdf}")
print(f"Saved bar plot -> {fig_png}")
plt.show()
```
