# Dynamic Neighborhood — ESS Analysis

This notebook analyzes the **ESS dynamics** in the evolving-neighborhood model.

Workflow:

1. Set up imports, paths, and plotting defaults
2. Run one-parameter sweeps for the evolving-neighborhood ESS model
3. Generate summary plots of final values $(N^*, u^*, y^*)$
4. Perform local central-difference sensitivity analysis around a baseline configuration

The code is organized to avoid repeated setup and repeated plotting logic.

## Sweep helpers

The helper functions below:

- enforce non-negativity
- run a single evolving-neighborhood ESS simulation
- save full trajectories to Excel
- generate diagnostic trajectory plots
- extract final values for compact summary plots


```python

# ============================================================
# Helper functions for evolving-neighborhood ESS sweeps
# ============================================================
def _tag(v) -> str:
    """Convert numeric values into filename-safe tags."""
    if isinstance(v, (int, np.integer)):
        return str(int(v))
    s = f"{float(v):.6g}"
    return s.replace("-", "m").replace(".", "p")


def _untag_value(token: str) -> float:
    """Convert a filename-safe tag back to a numeric value."""
    s = token.replace("p", ".")
    if s.startswith("m"):
        s = "-" + s[1:]
    return float(s)


def _truncate_at_first_negative(arr: np.ndarray) -> np.ndarray:
    """Set all values to zero after the first negative entry."""
    arr = np.asarray(arr, dtype=float)
    neg = arr < 0
    if neg.any():
        k = int(np.argmax(neg))
        out = arr.copy()
        out[k:] = 0.0
        return out
    return arr


def _enforce_nonneg(N: np.ndarray, u: np.ndarray, y: np.ndarray, mode: str = "truncate"):
    """Enforce non-negativity on the evolving-neighborhood state variables."""
    N = np.maximum(N, 0.0)
    if mode == "truncate":
        u = _truncate_at_first_negative(np.maximum(u, 0.0))
        y = _truncate_at_first_negative(np.maximum(y, 0.0))
    elif mode == "clip":
        u = np.maximum(u, 0.0)
        y = np.maximum(y, 0.0)
    else:
        raise ValueError("nonneg_mode must be 'truncate' or 'clip'")
    return N, u, y


def _run_one_ess(p: EvolvingParams, q: float, N0: float, u0: float, y0: float, nonneg_mode: str = "truncate"):
    """Run one evolving-neighborhood ESS simulation."""
    out = integrate_ess(q=q, p=p, N0=N0, u0=u0, y0=y0)

    if not out.get("success", True):
        print(f"[warn] LSODA: {out.get('message','')}")

    t = np.asarray(out["t"], dtype=float)
    N = np.asarray(out["N"], dtype=float)
    u = np.asarray(out["u"], dtype=float)
    y = np.asarray(out["y"], dtype=float)

    N, u, y = _enforce_nonneg(N, u, y, mode=nonneg_mode)
    return t, N, u, y


def _plot_three_panel_ess(t, N, u, y, p: EvolvingParams, q_used: float, title_extra: str,
                          figpath_base: Optional[Path] = None):
    """Diagnostic 3-panel trajectory plot for one ESS run."""
    fig, axs = plt.subplots(1, 3, figsize=FIGSIZE_RUN)

    axs[0].plot(t, N, linewidth=LINEWIDTH, color=COLOR_N)
    axs[0].set_title("N(t)")
    axs[0].set_xlabel("Time")
    axs[0].set_ylabel("N")
    axs[0].grid(True, alpha=0.25)

    axs[1].plot(t, u, linewidth=LINEWIDTH, color=COLOR_U)
    axs[1].set_title("u(t) — ESS")
    axs[1].set_xlabel("Time")
    axs[1].set_ylabel("u")
    axs[1].grid(True, alpha=0.25)

    axs[2].plot(t, y, linewidth=LINEWIDTH, color=COLOR_Y)
    axs[2].set_title("y(t) — ESS")
    axs[2].set_xlabel("Time")
    axs[2].set_ylabel("y")
    axs[2].grid(True, alpha=0.25)

    fig.tight_layout(rect=[0, 0, 1, 0.92])
    fig.suptitle(
        f"ESS (evolving N) — {title_extra} | q={q_used:.3g} | "
        f"R={p.R}, a={p.a}, k={p.k}, c={p.c}, h={p.h}, f={p.f}, "
        f"σ_u={p.sigma_u}, σ_y={p.sigma_y}, t_max={int(p.t_max)}",
        y=0.99,
        fontsize=16,
    )

    if figpath_base is not None:
        fig.savefig(figpath_base.with_suffix(".svg"))
        fig.savefig(figpath_base.with_suffix(".pdf"))

    plt.show()


def sweep_ess_evolving(
    param: str,
    values: Sequence[float],
    q_fixed: float = q_default,
    N0_init: float = N0_default,
    u0: float = u0_default,
    y0: float = y0_default,
    b_override: Optional[float] = None,
    sigma_u_override: Optional[float] = None,
    sigma_y_override: Optional[float] = None,
    tmax_override: Optional[float] = None,
    nonneg_mode: str = "truncate",
):
    """
    Sweep one parameter for the evolving-neighborhood ESS model.

    `param` can be:
    - "q", which is treated separately from the parameter dataclass
    - any field in EvolvingParams, such as k, a, R, c, h, or f
    """
    assert (param == "q") or hasattr(BASE_P, param), f"Unknown parameter '{param}'"

    outdir = SWEEP_DIR / str(param)
    outdir.mkdir(parents=True, exist_ok=True)

    excel_path = outdir / f"ess_evo_sweep_{param}_q{_tag(q_fixed)}_N0{_tag(N0_init)}.xlsx"
    print(f"Running ESS (evolving N) sweep over {param}, saving to {excel_path}")

    all_runs = {}

    with pd.ExcelWriter(excel_path, engine="openpyxl") as writer:
        pd.DataFrame({
            "Parameter": [param],
            "Values": [list(map(float, values))],
            "q_fixed": [q_fixed],
            "N0_init": [N0_init],
            "Overrides": [dict(
                b=b_override,
                sigma_u=sigma_u_override,
                sigma_y=sigma_y_override,
                t_max=tmax_override,
            )],
            "ICs": [dict(u0=u0, y0=y0)],
            "nonneg_mode": [nonneg_mode],
        }).to_excel(writer, sheet_name="params", index=False)

        for val in values:
            val = float(val)

            # If sweeping q, update q directly; otherwise q remains fixed
            q_used = val if param == "q" else q_fixed

            kwargs = {} if param == "q" else {param: val}
            if b_override is not None:
                kwargs["b"] = b_override
            if sigma_u_override is not None:
                kwargs["sigma_u"] = sigma_u_override
            if sigma_y_override is not None:
                kwargs["sigma_y"] = sigma_y_override
            if tmax_override is not None:
                kwargs["t_max"] = tmax_override

            p_run = replace(BASE_P, **kwargs)

            # Run the simulation
            t, N, u, y = _run_one_ess(p_run, q_used, N0_init, u0, y0, nonneg_mode=nonneg_mode)

            # Save trajectory to workbook
            df = pd.DataFrame({"t": t, "N": N, "u": u, "y": y})
            sheet_name = f"{param}_{_tag(val)}"
            df.to_excel(writer, sheet_name=sheet_name, index=False)
            all_runs[sheet_name] = df

            # Diagnostic figure for this run
            title_extra = f"{param}={val:.3g}" if param != "q" else f"q={val:.3g}"
            figbase = outdir / f"{sheet_name}_diagnostic"
            _plot_three_panel_ess(t, N, u, y, p_run, q_used=q_used, title_extra=title_extra, figpath_base=figbase)

    print(f"Finished sweep: {param}")
    return all_runs, excel_path


def extract_final_ess_evolving(one_sweep_result):
    """Extract final values from one in-memory sweep result dictionary."""
    rows = []

    for sheet_name, df in one_sweep_result.items():
        if df.empty:
            continue

        try:
            token = sheet_name.split("_", 1)[1]
            param_value = _untag_value(token)
        except Exception:
            continue

        rows.append({
            "parameter_value": param_value,
            "N_final": float(df["N"].iloc[-1]),
            "u_final": float(df["u"].iloc[-1]),
            "y_final": float(df["y"].iloc[-1]),
            "t_final": float(df["t"].iloc[-1]),
        })

    return pd.DataFrame(rows).sort_values("parameter_value").reset_index(drop=True)


def plot_ess_evolving_summary(df_final, param_name, xlabel=None, save=True):
    """
    Plot the final sweep summary with two equal-sized subplots:

    - Left: N_ESS vs parameter
    - Right: u_ESS and y_ESS vs parameter on twin y-axes
    """
    if df_final.empty:
        print(f"[warn] No data available for {param_name}")
        return

    x = df_final["parameter_value"].values
    Nf = df_final["N_final"].values
    uf = df_final["u_final"].values
    yf = df_final["y_final"].values

    fig, axs = plt.subplots(1, 2, figsize=(10.5, 3.8))
    axN = axs[0]
    ax1 = axs[1]
    ax2 = ax1.twinx()

    # Left panel: final N
    axN.plot(x, Nf, linewidth=LINEWIDTH, color=COLOR_N)
    axN.set_xlabel(xlabel if xlabel else param_name, fontsize=12)
    axN.set_ylabel(r"$N_{\mathrm{ESS}}$", fontsize=14)
    axN.tick_params(axis="x", labelsize=11)
    axN.tick_params(axis="y", labelsize=11, colors="black")
    axN.grid(True, alpha=0.25)

    # Right panel: final u and y
    line1 = ax1.plot(x, uf, linewidth=LINEWIDTH, color=COLOR_U, label=r"$u_{\mathrm{ESS}}$")
    line2 = ax2.plot(x, yf, linewidth=LINEWIDTH, color=COLOR_Y, label=r"$y_{\mathrm{ESS}}$")

    ax1.set_xlabel(xlabel if xlabel else param_name, fontsize=12)
    ax1.set_ylabel(r"$u_{\mathrm{ESS}}$", color=COLOR_U, fontsize=14)
    ax2.set_ylabel(r"$y_{\mathrm{ESS}}$", color=COLOR_Y, fontsize=14)

    ax1.tick_params(axis="x", labelsize=11)
    ax1.tick_params(axis="y", labelsize=11, colors=COLOR_U)
    ax2.tick_params(axis="y", labelsize=11, colors=COLOR_Y)

    ax1.spines["left"].set_color(COLOR_U)
    ax2.spines["right"].set_color(COLOR_Y)
    ax1.grid(True, alpha=0.25, color="gray")

    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc="upper right", frameon=False, fontsize=11)

    fig.subplots_adjust(left=0.08, right=0.92, bottom=0.18, top=0.94, wspace=0.35)

    if save:
        fig.savefig(SUMMARY_DIR / f"ess_evo_summary_{param_name}.png", bbox_inches="tight")
        fig.savefig(SUMMARY_DIR / f"ess_evo_summary_{param_name}.svg", bbox_inches="tight")

    plt.show()
```

# Sensitivity analysis

We compute local central-difference sensitivities of the evolving-neighborhood ESS equilibrium values
$(N^*, u^*, y^*)$ with respect to the parameters
$$
p \in \{q, k, a, R, c, h, f\}.
$$

Here $N_0$ is treated as an initial condition rather than a model parameter, so it is not included in the local sensitivity set.


## Central-difference perturbations

Let $\alpha = \mathrm{FRAC\_STEP}$ and per-parameter floors $\mathrm{MIN\_ABS}[p]$. For each parameter $p$,
$$
\Delta p = \max\!\big(\alpha |p_0|,\ \mathrm{MIN\_ABS}[p]\big), \qquad
p_- = p_0 - \Delta p, \qquad
p_+ = p_0 + \Delta p.
$$

We estimate local derivatives by central differences and convert them to normalized sensitivities for comparison across parameters.


```python

# ============================================================
# Sensitivity analysis setup
# ============================================================
BASE_SENS = replace(
    EvolvingParams(),
    t_max=50_000_000.0,
    sigma_u=0.9,
    sigma_y=0.9,
    b=1.5,
)

# Baseline initial condition and baseline q for local sensitivity analysis
N0_sens = 5.0
u0_sens = 300.0
y0_sens = 50.0
q0_sens = 0.5

params_of_interest = ["q", "k", "a", "R", "c", "h", "f"]

FRAC_STEP = 0.10
MIN_ABS = {
    "q": 1e-2,
    "k": 1e-3,
    "a": 1e-3,
    "R": 0.5,
    "c": 1e-3,
    "h": 1e-3,
    "f": 1e-3,
}
EPS = 1e-9
```

## Sensitivity helpers

These functions:

- integrate the ESS dynamics until equilibrium
- extend $t_{\max}$ adaptively if needed
- compute the baseline equilibrium
- compute local central-difference sensitivities
- generate diagnostic dynamics plots


```python

# ============================================================
# Helper functions for local sensitivity analysis
# ============================================================
def has_equilibrated(t, N, u, y, window_frac=0.1, atol=1e-5, rtol=1e-4):
    """Check whether N, u, and y have stabilized near the end of the trajectory."""
    n = len(t)
    if n < 5:
        return False

    k0 = max(0, int((1.0 - window_frac) * n))

    def _ok(arr):
        arr_w = arr[k0:]
        a_f = arr_w[-1]
        return np.max(np.abs(arr_w - a_f)) <= (atol + rtol * max(1.0, abs(a_f)))

    return bool(_ok(N) and _ok(u) and _ok(y))


def integrate_to_equilibrium_ess(
    p: EvolvingParams,
    q: float,
    N0: float,
    u0: float,
    y0: float,
    max_ext: int = 4,
    tmax_factor: float = 2.0,
    check_kwargs: Optional[dict] = None,
    nonneg_mode: str = "truncate",
):
    """Integrate ESS dynamics, extending t_max until equilibration or the extension limit."""
    if check_kwargs is None:
        check_kwargs = dict(window_frac=0.1, atol=1e-5, rtol=1e-4)

    p_run = p
    ext = 0

    while True:
        out = integrate_ess(q=q, p=p_run, N0=N0, u0=u0, y0=y0)

        if not out.get("success", True):
            print(f"[warn] LSODA: {out.get('message','')}")

        t = np.asarray(out["t"], dtype=float)
        N = np.asarray(out["N"], dtype=float)
        u = np.asarray(out["u"], dtype=float)
        y = np.asarray(out["y"], dtype=float)

        N, u, y = _enforce_nonneg(N, u, y, mode=nonneg_mode)

        if has_equilibrated(t, N, u, y, **check_kwargs):
            return {
                "t": t,
                "N": N,
                "u": u,
                "y": y,
                "p_used": p_run,
                "q_used": q,
                "converged": True,
                "extensions": ext,
            }

        if ext >= max_ext:
            return {
                "t": t,
                "N": N,
                "u": u,
                "y": y,
                "p_used": p_run,
                "q_used": q,
                "converged": False,
                "extensions": ext,
            }

        p_run = replace(p_run, t_max=p_run.t_max * tmax_factor)
        ext += 1


def _compute_G(N_star, u_star, y_star, q, p_used):
    """Call G robustly across signatures that may or may not expose q explicitly."""
    try:
        return float(G(N_star, u_star, y_star, q=q, p=p_used))
    except TypeError:
        return float(G(N_star, u_star, y_star, p=p_used))


def _run_eq(p: EvolvingParams, q: float):
    """Compute a full equilibrium object for one (p, q) combination."""
    run = integrate_to_equilibrium_ess(
        p=p,
        q=q,
        N0=N0_sens,
        u0=u0_sens,
        y0=y0_sens,
        max_ext=4,
        tmax_factor=2.0,
        check_kwargs=dict(window_frac=0.1, atol=1e-5, rtol=1e-4),
        nonneg_mode="truncate",
    )

    N_star = float(run["N"][-1])
    u_star = float(run["u"][-1])
    y_star = float(run["y"][-1])
    G_star = _compute_G(N_star, u_star, y_star, q=run["q_used"], p_used=run["p_used"])

    return {"N_star": N_star, "u_star": u_star, "y_star": y_star, "G_star": G_star, "run": run}


def _perturb(prm: str, base_p: EvolvingParams, base_q: float, frac=FRAC_STEP):
    """
    Return (q-, p-), (q+, p+) pairs for central differences.
    q is external to the parameter dataclass.
    """
    if prm == "q":
        q = float(base_q)
        dq = max(abs(frac * q), MIN_ABS["q"])
        qm = max(q - dq, 0.0)
        qp = min(q + dq, 1.0)
        return (qm, base_p), (qp, base_p)

    p0 = float(getattr(base_p, prm))
    dp = max(abs(frac * p0), MIN_ABS[prm])
    pm = replace(base_p, **{prm: max(p0 - dp, 0.0)})
    pp = replace(base_p, **{prm: p0 + dp})
    return (base_q, pm), (base_q, pp)


def plot_sensitivity_diagnostics(runs: dict, title: str, save_path: Optional[Path] = None):
    """Plot baseline, minus, and plus dynamics for N(t), u(t), and y(t)."""
    fig, axs = plt.subplots(1, 3, figsize=(13, 4))

    for lab, (t, N, u, y) in runs.items():
        axs[0].plot(t, N, linewidth=2, label=lab)
        axs[1].plot(t, u, linewidth=2, label=lab)
        axs[2].plot(t, y, linewidth=2, label=lab)

    axs[0].set_xlabel("time")
    axs[0].set_ylabel("N(t)")
    axs[1].set_xlabel("time")
    axs[1].set_ylabel("u(t)")
    axs[2].set_xlabel("time")
    axs[2].set_ylabel("y(t)")

    for ax in axs:
        ax.grid(True, alpha=0.25)

    axs[0].legend(frameon=False, fontsize=9)
    fig.suptitle(title, y=0.98, fontsize=16)
    fig.tight_layout(rect=[0, 0, 1, 0.95])

    if save_path is not None:
        fig.savefig(save_path.with_suffix(".pdf"))
        fig.savefig(save_path.with_suffix(".png"), dpi=300)

    plt.show()
```

## Run the local sensitivity analysis

The next cell computes:

- the baseline evolving-neighborhood ESS equilibrium
- local normalized sensitivities for $N^*$, $u^*$, $y^*$, and $G^*$
- diagnostic baseline / minus / plus dynamics for each parameter
- a grouped bar plot of $S_N$, $S_u$, and $S_y$


```python

# ============================================================
# Baseline equilibrium
# ============================================================
base_full = _run_eq(BASE_SENS, q0_sens)
base_eq = {k: base_full[k] for k in ("N_star", "u_star", "y_star", "G_star")}

print(
    f"[baseline ESS] q={q0_sens} | "
    f"N*={base_eq['N_star']:.4g}, "
    f"u*={base_eq['u_star']:.4g}, "
    f"y*={base_eq['y_star']:.4g}, "
    f"G*={base_eq['G_star']:.4g}, "
    f"t_max_used={int(base_full['run']['p_used'].t_max)}"
)

# ============================================================
# Local central-difference sensitivities
# ============================================================
rows = []

for prm in params_of_interest:
    # Build minus / plus perturbations
    (qm, pm), (qp, pp) = _perturb(prm, BASE_SENS, q0_sens, frac=FRAC_STEP)

    # Equilibria under minus / plus perturbations
    eq_m = _run_eq(pm, qm)
    eq_p = _run_eq(pp, qp)

    # Actual perturbation size and baseline parameter value
    if prm == "q":
        dp = qp - qm
        p0_val = q0_sens
    else:
        dp = getattr(pp, prm) - getattr(pm, prm)
        p0_val = getattr(BASE_SENS, prm)

    # Central-difference derivatives
    dN_dp = (eq_p["N_star"] - eq_m["N_star"]) / dp
    du_dp = (eq_p["u_star"] - eq_m["u_star"]) / dp
    dy_dp = (eq_p["y_star"] - eq_m["y_star"]) / dp
    dG_dp = (eq_p["G_star"] - eq_m["G_star"]) / dp

    # Normalized sensitivities
    SN = (p0_val / max(abs(base_eq["N_star"]), EPS)) * dN_dp
    Su = (p0_val / max(abs(base_eq["u_star"]), EPS)) * du_dp
    Sy = (p0_val / max(abs(base_eq["y_star"]), EPS)) * dy_dp
    SG = (p0_val / max(abs(base_eq["G_star"]), EPS)) * dG_dp

    rows.append({
        "param": prm,
        "p0": p0_val,
        "dp": dp,
        "q0": q0_sens,
        "N_star": base_eq["N_star"],
        "u_star": base_eq["u_star"],
        "y_star": base_eq["y_star"],
        "G_star": base_eq["G_star"],
        "N_minus": eq_m["N_star"],
        "N_plus": eq_p["N_star"],
        "dN/dp": dN_dp,
        "S_N": SN,
        "u_minus": eq_m["u_star"],
        "u_plus": eq_p["u_star"],
        "du/dp": du_dp,
        "S_u": Su,
        "y_minus": eq_m["y_star"],
        "y_plus": eq_p["y_star"],
        "dy/dp": dy_dp,
        "S_y": Sy,
        "G_minus": eq_m["G_star"],
        "G_plus": eq_p["G_star"],
        "dG/dp": dG_dp,
        "S_G": SG,
        "tmax_used_minus": int(eq_m["run"]["p_used"].t_max),
        "tmax_used_plus": int(eq_p["run"]["p_used"].t_max),
    })

sens_df = pd.DataFrame(rows).set_index("param")

print("\n[summary] ESS (evolving N) local sensitivities:")
display_cols = [
    "S_N", "S_u", "S_y", "S_G",
    "dN/dp", "du/dp", "dy/dp", "dG/dp",
    "p0", "dp", "tmax_used_minus", "tmax_used_plus"
]
display(sens_df[display_cols])

# ============================================================
# Save sensitivity tables
# ============================================================
sens_xlsx = SENS_DIR / "ess_evo_local_sensitivity.xlsx"
with pd.ExcelWriter(sens_xlsx, engine="openpyxl", mode="w") as writer:
    pd.DataFrame([{
        "N0": N0_sens,
        "u0": u0_sens,
        "y0": y0_sens,
        "q0": q0_sens,
        "t_max": BASE_SENS.t_max,
        "atol": BASE_SENS.atol,
        "rtol": BASE_SENS.rtol,
        "b": BASE_SENS.b,
        "R": BASE_SENS.R,
        "a": BASE_SENS.a,
        "k": BASE_SENS.k,
        "c": BASE_SENS.c,
        "h": BASE_SENS.h,
        "f": BASE_SENS.f,
        "sigma_u": BASE_SENS.sigma_u,
        "sigma_y": BASE_SENS.sigma_y,
        "frac_step": FRAC_STEP,
    }]).to_excel(writer, sheet_name="params", index=False)

    pd.DataFrame([base_eq]).to_excel(writer, sheet_name="baseline_eq", index=False)
    sens_df.reset_index().to_excel(writer, sheet_name="local_sensitivity", index=False)

print(f"Saved local sensitivity Excel -> {sens_xlsx}")

# ============================================================
# Per-parameter diagnostic trajectories
# ============================================================
for prm in params_of_interest:
    (qm, pm), (qp, pp) = _perturb(prm, BASE_SENS, q0_sens, frac=FRAC_STEP)

    run_b = base_full["run"]
    run_m = integrate_to_equilibrium_ess(pm, qm, N0_sens, u0_sens, y0_sens, nonneg_mode="truncate")
    run_p = integrate_to_equilibrium_ess(pp, qp, N0_sens, u0_sens, y0_sens, nonneg_mode="truncate")

    runs = {
        f"baseline (q={q0_sens}, t_max={int(run_b['p_used'].t_max)})": (run_b["t"], run_b["N"], run_b["u"], run_b["y"]),
        f"{prm}− (q={qm:.3g}, t_max={int(run_m['p_used'].t_max)})": (run_m["t"], run_m["N"], run_m["u"], run_m["y"]),
        f"{prm}+ (q={qp:.3g}, t_max={int(run_p['p_used'].t_max)})": (run_p["t"], run_p["N"], run_p["u"], run_p["y"]),
    }

    plot_sensitivity_diagnostics(
        runs,
        title=f"ESS (evolving N) — {prm} sensitivity",
        save_path=SENS_DIAG_DIR / f"ess_evo_sens_{prm}",
    )

# ============================================================
# Grouped bar plot of normalized sensitivities
# ============================================================
fig, ax = plt.subplots(figsize=(10.5, 5.0))
x = np.arange(len(params_of_interest))
width = 0.25

S_N_vals = [sens_df.loc[p, "S_N"] for p in params_of_interest]
S_u_vals = [sens_df.loc[p, "S_u"] for p in params_of_interest]
S_y_vals = [sens_df.loc[p, "S_y"] for p in params_of_interest]

ax.bar(x - width, S_N_vals, width, label=r"$N_{\mathrm{ESS}}^*$", color=COLOR_N)
ax.bar(x,         S_u_vals, width, label=r"$u_{\mathrm{ESS}}^*$", color=COLOR_U)
ax.bar(x + width, S_y_vals, width, label=r"$y_{\mathrm{ESS}}^*$", color=COLOR_Y)

ax.axhline(0, linewidth=1, color="k")
ax.set_xticks(x)
ax.set_xticklabels(params_of_interest)
ax.set_ylabel("Sensitivity")
ax.set_title(f"ESS local sensitivities (evolving N) | baseline q={q0_sens}")
ax.legend(frameon=False)
ax.grid(False)

fig.tight_layout()

sens_bar_pdf = SENS_DIR / "ess_evo_local_sensitivity_bars.pdf"
sens_bar_png = SENS_DIR / "ess_evo_local_sensitivity_bars.png"
fig.savefig(sens_bar_pdf)
fig.savefig(sens_bar_png, dpi=300)

print(f"Saved bar plot -> {sens_bar_pdf}")
print(f"Saved bar plot -> {sens_bar_png}")

plt.show()
```
